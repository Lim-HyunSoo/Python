```python
from bs4 import BeautifulSoup as bs
# beautifulsoup 사용목적 시험 
```

```python
h = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/98.0.4758.102 Safari/537.36'}
```

```python
res = req.get('https://www.melon.com/chart/index.htm', headers = h)
```

```python
soup = bs(res.text, 'lxml')
```

```python
song_info = soup.select("div.wrap_song_info")
```

```python
song_info[4].text
```

```python
info_len = len(song_info)
n = '\n\n\n'
cnt = 1
for ind in range(0, info_len, 2):
    info = song_info[ind].text
    
    print(cnt , ':', info[2:info.find(n)])
    cnt += 1
```

```python
# 클래스값에 공백이 있는 경우 > 공백을 . 으로 대체한다.
singer_info = soup.select("div.ellipsis.rank02")
```

```python
# 순수한 글자 정보만 담겨져있는 리스트 제작 
singerList = []
songList = []
rankList = []

for i in range(len(song_info2)):
    singerList.append(singer_info2[i].text.strip())
    songList.append(song_info2[i].text.strip())
    rankList.append(i+1)
```

```python
import pandas as pd
```

```python
info = {'가수' : singerList, '노래' : songList, '순위' : rankList}
```

```python
melon = pd.DataFrame(info)
```

```python
melon
```

```python
melon.set_index("순위", inplace = True)
```

```python
melon
```

```python
# 한글 인코딩 방법
# 1. euc-kr
# 2. ""
# 3. utf-8-sig
melon.to_csv("멜론차트.csv", encoding = "euc-kr")
```
