# 네이버 메인에서 글자정보 수집

```python
res = req.get("http://www.naver.com")
```

```python
from bs4 import BeautifulSoup as bs
```

```python
soup = bs(res.text, "lxml")
```

```python
# select의 리턴값은 리스트 형태 
naver = soup.select("a.nav")
```

```python
naver[0].text
```

```python
for i in naver:
    print(i.text)
```

```python
naver2 = soup.select_one("a.nav")
```

```python
naver2
```

# 뉴스 타이틀만 수집

```python
res2 = req.get("https://search.naver.com/search.naver?where=news&sm=tab_jum&query=%EC%9A%B0%ED%81%AC%EB%9D%BC%EC%9D%B4%EB%82%98")
```

```python
soup2 = bs(res2.text, "lxml")
```

```python
news_title = soup2.select("a.news_tit")
```

```python
for t in news_title:
    print(t.text)
```
