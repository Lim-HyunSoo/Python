```python
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
#1. req를 통해서 영화정보사이트 요청
res = req.get("https://movie.naver.com/movie/bi/mi/pointWriteFormList.naver?code=154282&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false")
```

```python
#2. 응답코드를 확인하고 정상적으로 통신이 됬다면
# bs데이터로 변형!
soup = bs(res.text,"lxml")
```

```python
# 관람객이 남긴 리뷰만 수집 * select()
review = soup.select("div.score_reple > p")
```

```python
# 리뷰데이터에 바로 접근이 불가능 이유는 실제 데이터가 다른곳에 저장
# <iframe> = 다른 서버 공간에 있는 데이터를 빌려다가 보여줄 수 있게 만들어주는 태그
# f12개발자 도구에서 ctrl+f(찾기기능) / iframe검색
# 실제 사용되고있는 iframe 찾고, 실제 데이터의 경로인 src주소를 복사
# 메인서버url(http://movie.naver.com) + src경로 합쳐서 접근
# 다시 새로운 URL로 요청
```

```python
for i in review :
    print(i.text.strip())
```

```python
# 관람객 아이콘 삭제
# 불필요한 요소를 제거하는 함수
# 불필요한 데이터를 수집(컴퓨터가 알아먹을 수 있게)
icon = soup.select("span.ico_viewer")
```

```python
# 제거하기 위한 함수 extract()
for i in icon :
    i.extract()
```

```python
for i in review :
    print(i.text.strip())
```

# 1페이지부터 10페이지까지 리뷰데이터를 수집

```python
# 데이터를 요청, 변경, 수집하고, 출력 총 10번을 반복
# 반복문 작성 팁!
# for부터 적는게 아니고! 내가 반복할 코드들을 먼저 한번 작성해서 완벽한 로직을 제작
# 하나의 완벽한 싸이클을 제작 후 for로 감싸주기
# 감싼 후, i값의 활용방안을 생각!
for i in tq(range(1,51,1)) :
    time.sleep(1)
    # 요청하는 페이지의 번호가 바뀌니까 마지막에 i값을 추가
    # i는 정수형데이터이기 때문에 str()문자형으로 형변환 필요!
    res = req.get("https://movie.naver.com/movie/bi/mi/pointWriteFormList.naver?code=154282&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page="+str(i))
    soup = bs(res.text,"lxml")
    icon = soup.select("span.ico_viewer")
    for i in icon :
        i.extract()
    review = soup.select("div.score_reple > p")
    for i in review :
        print(i.text.strip())
```
