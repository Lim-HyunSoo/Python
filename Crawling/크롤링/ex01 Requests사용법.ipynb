# Requests 라이브러리
- 내가 원하는 페이지의 정보를 요청하는 라이브러리
- 브라우저의 역할을 대신

```python
import requests as req
```

```python
# req야 네이버메인페이지 정보좀 가지고와줘!
# 1. 페이지 정보를 요청할때는 req.get("접근하고자 하는 페이지의 URL")
# 2. response[200] = 통신에 성공했습니다.
res = req.get("http://www.naver.com")
```

```python
res
```

```python
# 넘겨받은 response객체에서 순수한 html글자정보를 보는 방법 .text
res.text
```

# beautifulSoup 라이브러리
## bs사용 목적 문제 ★

```python
# from bs4 키워드를 통해서 접근
# stonsoup이 아닌 soup을 임포트!
from bs4 import BeautifulSoup as bs
```

```python
#bs("내가 변형할 텍스트데이터", "파싱방법 = lxml")
# bs데이터 가공시키는 방법 문제 ★
soup = bs(res.text,"lxml")
```

```python
soup
```

```python
#.select("선택자") > soup데이터에서 내가 필요한 요소를 추출
# 1. select("a") > 내가보고있는 soup데이터에서 모든 a태그를 가지고와라!
# 2. 공통적으로 가지고있는 구분자를 탐색! class값이 nav인 요소들
# 3. select("a.nav") > 내가보고있는 soup데이터에서 a태그들 중에 클래스값이 nav요소들을 가지고와라!
# 4. select의 리턴데이터타입은 리스트 타입!
naver = soup.select("a.nav")
```

```python
# 특정 요소(태그+컨텐츠)에서 글자만 꺼내올때는 .text
# bs데이터 가공시킨데이터 글자출력 문제 ★
naver[0].text
```

```python
naver[1].text
```

```python
naver[2].text
```

```python
for i in naver :
    print(i.text)
```

```python
# select_one("선택자") > 모든 데이터에서 첫번째 데이터만 가지고오겠다!
# select vs select_one 차이점 문제★
naver2 = soup.select_one("a.nav")
```

```python
naver2.text
```

# 뉴스 타이틀만 수집

```python
res = req.get("https://search.naver.com/search.naver?where=news&query=%EC%9A%B0%ED%81%AC%EB%9D%BC%EC%9D%B4%EB%82%98&sm=tab_tmr&nso=so:r,p:all,a:all&sort=0")
```

```python
soup = bs(res.text,"lxml")
```

```python
news = soup.select("a.news_tit")
```

```python
for i in news :
    print(i.text)
```
