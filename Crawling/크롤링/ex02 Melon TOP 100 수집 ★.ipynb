```python
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
#1. 멜론차트 100사이트정보를 요청 (url)값을 통해서
req.get('https://www.melon.com/chart/index.htm')
```

```python
#2. 응답코드가 200번대가 잘 나오는지 체크!
#2.1 혹시 응답코드가 400번대 인경우는 head작업을 통해서 데이터를 받아오기!
h = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/98.0.4758.102 Safari/537.36'}
res = req.get('https://www.melon.com/chart/index.htm',headers = h)
res.text
```

```python
#3. 컴퓨터가 데이터에 접근할수 있게 bs데이터 변형(html문서화)
soup = bs(res.text,"lxml")
```

```python
#4. 가수 이름을 수집
# * 가수정보를 가지고있는 a태그를 접근함!
# * a태그에 구분자가 존재하지 않아서 데이터를 구분지을 수가 없다!
# * 이런경우에는 바로 위에 존재하는 부모태그를 검사.
# * 부모태그 또한 구분자가 필요함! 없는경우에는 그 부모태그를 다시 검사
# * 클래스값이 중간에 공백이 있는 경우 = 클래스값을 여러개를 가지고있다.
# * 해결방법 = 공백을 지우고 .통해서 값을 이어줘야한다!
singer = soup.select("div.ellipsis.rank02 > a")
```

```python
# 데이터의 개수를 확인해봤더니 117개!
# 가수가 여러명인 경우 따로따로 데이터를 수집
len(singer)
```

```python
# a태그를 묶음으로 가지고올 수 있게 부모태그 통으로 수집
singer = soup.select("div.ellipsis.rank02")
# 길이가 100개는 잘 출력
len(singer)
# 데이터가 중복이 된걸 확인!
for i in singer :
    print(i.text)
```

```python
# 마지막으로 확인해보지 않았던 span태그로 접근
newSinger = soup.select("span.checkEllipsis")
len(newSinger)
for i in newSinger :
    print(i.text)
```

```python
# 노래제목 수집!
song = soup.select("div.ellipsis.rank01 > span")
```

```python
for i in song :
    print(i.text.strip())
```

```python
# 불필요한 개행문자를 제거하는 방법
song[0].text.strip()
```

```python
# 순수한 글자 정보만 담겨져있는 리스트 제작
# 내가 수집한 newSinger, song리스트는 태그정보가 포한됨 데이터(컴퓨터를 위한)
singerList = []
songList = []
rankList = []
for i in range(len(song)) :
    singerList.append(newSinger[i].text.strip())
    songList.append(song[i].text.strip())
    rankList.append(i+1)
```

```python
# 데이터프레임을 제작
import pandas as pd
```

```python
# 딕셔너리형태의 데이터를 제작(키와 벨류)
info = {"가수" : singerList, "노래":songList, "순위" : rankList}
```

```python
melon = pd.DataFrame(info)
```

```python
# 내가 원하는 컬럼으로 인덱스값을 대체
melon.set_index("순위",inplace = True)
```
